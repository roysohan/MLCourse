```python
sn.kdeplot( data = chd_df, x = 'tobacco');
```

```python
yes_chd = chd_df[chd_df.chd ==1]
no_chd = chd_df[chd_df.chd == 0]
```

```python
sn.kdeplot(data = yes_chd, x='tobacco', label = 'CHD')
sn.kdeplot(data = no_chd, x ='tobacco', label = 'NO CHD')
```

```python
yes_chd.tobacco.mean()
```

```python
no_chd.tobacco.mean()
```
